#### <font color ="purple">___Read in CSV(s) into dataframe object for capturing plotable queries</font>

```python
class covidWorker():
    def __init__(self,daycount,datasetlabel):
        #so far the CSVs have had the same layout so boiling out timeframes desired
        
        self.descriptor = datasetlabel
        self.workingdate = datetime.today().strftime("%m/%d/%y")
        self.daysBack = daycount
        self.deathDF = pd.read_csv(path + "time_series_19-covid-Deaths.csv")
        self.confirmDF = pd.read_csv(path + "time_series_19-covid-Confirmed.csv")
        self.recoverDF = pd.read_csv(path + "time_series_19-covid-Recovered.csv")
        self.lastgroupby = "NA"
        self.pltgridsize = 18, 9
        self.pltkind = "area"
        self.pltcmap = "Dark2"
        self.xaxis ="xdata"
        self.Yaxis ="ydata"
        self.wrkDFcolcount = 0
        self.wrkDF=self.setWORKERDF("death")
        
        
      
  
   
    def __repr__(self):
        return self.descriptor + "has been set"
           
    def getGroupBySum(self,groupby):
        try:
                self.lastgroupby = groupby

          
                self.wrkDF = self.wrkDF.sum(groupby)
                self.wrkDF = self.wrkDF.iloc[:,(self.wrkDFcolcount - self.daysBack): self.wrkDFcolcount]
                self.wrkDF.sort_values(by=self.wrkDF[self.wrkDFcolcount-1].name,ascending=False,inplace=True)
                #self.wrkDF.reset_index(groupby,inplace=True)
                return self.wrkDF
            
        except:
            return groupby + " column does not exsist possibly ---Failure"

        
    def returnManualiloc(self,selectDF,colstart,colend):
        if selectDF == "death":
            return self.deathDF.iloc[:colstart:colend]
        if selectDF == "confirm":
            return self.confirmDF.iloc[:colstart:colend]
        if selectDF == "recover":
            return self.recoverDF.iloc[:colstart:colend]
        else:
            return False  
        
        
    def setWORKERDF(self,selectDF):
        if selectDF == "death":
            self.wrkDF = self.deathDF
            self.delcol(['Lat', 'Long'])
        if selectDF == "confirm":
            self.wrkDF = self.confirmDF
            self.delcol(['Lat', 'Long'])
        if selectDF == "recover":
            self.wrkDF = self.recoverDF
            self.delcol(['Lat', 'Long'])
        else:
            return False  
        
        
    def delcol(self,collist):
        try:
            self.wrkDF.drop(columns=collist,axis=1,inplace=True)
            self.wrkDFcolcount = len(self.wrkDF.columns)
        except:
            self.wrkDFcolcount = len(self.wrkDF.columns)
        
    def setPlotter(self,size,color,kind):
        self.pltgridsize = size
        self.pltkind = kind
        self.pltcmap = color
        
   
    def getPlot(self,title,transpose,headcount):
            
        if transpose == False:
            myPLT = self.wrkDF.head(headcount).plot(kind=self.pltkind,figsize =self.pltgridsize)
            myPLT.figure.suptitle(title)
            return myPLT
        else:
            myPLT = self.wrkDF.head(headcount).T.plot(kind=self.pltkind,figsize =self.pltgridsize)
            myPLT.figure.suptitle(title)
            return myPLT

       
        
    def savePLTs(self,plt,filename):
        with PdfPages(filename) as pdf:
            plt.figsize = (18,10)
            pdf.savefig(plt.figure)
            d = pdf.infodict()
            d['Title'] = 'COVID19_plot output'
            d['Author'] = 'MFRY'
            d['Subject'] = 'COVID19dataset'
            d['Keywords'] = 'COVID19 PLOT'
            d['CreationDate'] = datetime.today()
            d['ModDate'] = datetime.today()
            
```
